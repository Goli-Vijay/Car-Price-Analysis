```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python

brand= []
Type = []
seating = []
model = []
fuel_type=[]
gear_type=[]
price=[]
Location=[]
for i in range(1,11):
    page = requests.get(f"https://ackodrive.com/collection/express-delivery-cars/mumbai/page/{i}/")
    soup = BeautifulSoup(page.text)
    for i in soup.find_all('div',class_='styles__Wrapper-sc-8077a65d-0 fZOnVd'):
        rm= i.find('h2',class_='styles__ModelTitle-sc-aa3bd0b2-4 ikEBAq')
        if rm:
            b=re.findall('^[A-Za-z]+',rm.text)
            if b:
                brand.append(b[0])
            else:
                brand.append(np.nan)
                
            m = re.findall("^[A-Za-z]+\s(.+)",rm.text)
            if m:
                model.append(m[0])
            else:
                model.append(np.nan)
        p=i.find('div',class_='styles__Price-sc-aa3bd0b2-18 bzmdsn')
        if p:
            price.append(p.text)
        else:
            price.append(np.nan)
        reg = i.find('div','styles__VariantInfo-sc-aa3bd0b2-12 kCNdrj')
        if reg:
            f = re.findall('Petrol|Diesel|CNG|Electric',reg.text)
            if f:
                fuel_type .append(f[0])
            else:
                fuel_type.append(np.nan)
            g=re.findall('Manual|Automatic',reg.text)
            if g:
                gear_type.append(g[0])
            else:
                gear_type.append(np.nan)
        regex = i.find('div',class_='styles__ModelInfo-sc-aa3bd0b2-8 cEyrJI') 
        if regex:
            t = re.findall(r'^[A-Za-z]+',regex.text)
            if len(t)>0:
                Type.append(t[0])
            else:
                Type.append(np.nan)
            s = re.findall('\d\sSeater',regex.text)
            if s:
                seating.append(s[0])
            else:
                seating.append(np.nan)
            v = re.findall('Seater(.*)',regex.text)
        l = i.find('div',class_='styles__CityName-sc-aa3bd0b2-17 bysq')
        if l:
            lo = re.findall('in\s(.*)',l.text)
            if lo: 
                Location.append(lo[0])
            else:
                Location.append(np.nan)

    
```

```python
import pandas as pd
mumbai= pd.DataFrame({'Brand':brand,'Model':model,'Type':Type,'Seats':seating,'Price':price,'Fuel':fuel_type,'Gear_type':gear_type,'Location':Location})
```
